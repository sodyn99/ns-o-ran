```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd

carbon = pd.read_csv("./emissions.csv")

test = carbon[carbon['project_name'] == 'test']
centralized = carbon[carbon['project_name'] == 'centralized'].reset_index(drop=True)
federated = carbon[carbon['project_name'] == 'federated'].reset_index(drop=True)
```

```python
centralized = centralized.truncate(before=None,after=40,axis=0)
federated = federated.truncate(before=None,after=40,axis=0)
```

```python
mpl.rcdefaults()
mpl.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(figsize=(10,4))
plt.grid(True, linestyle='--')
plt.title('CO2 Emissions')
plt.plot(range(len(centralized['emissions_rate'])), centralized['emissions_rate'], label='Emssions Rate', marker="^", linestyle=":")#, color='red')
plt.plot(range(len(centralized['emissions'])), centralized['emissions'], label='Emissions', marker="", linestyle=":")#, color='red')
plt.xlabel('Episode')
# plt.ylabel('Time')
plt.legend(prop={'size': 12})

plt.savefig('plot_carbon-centralized.png', bbox_inches='tight')
plt.show()
```

```python
mpl.rcdefaults()
mpl.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(figsize=(10,4))
plt.grid(True, linestyle='--')
plt.title('CO2 Emissions Rate')
plt.plot(range(len(centralized['emissions_rate'])), centralized['emissions_rate'], label='Centralized', marker="", linestyle="-")#, color='red')
plt.plot(range(len(federated['emissions_rate'])), federated['emissions_rate'], label='Federated', marker="", linestyle="-")#, color='k')
plt.xlabel('Episode')
# plt.ylabel('Time')
plt.legend(prop={'size': 12})

plt.savefig('plot_carbon-emissions-rate.png', bbox_inches='tight')
plt.show()
```
